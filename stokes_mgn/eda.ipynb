```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from modulus.datapipes.gnn.utils import read_vtp_file
import pandas as pd
num_training_samples = 50
```

```python
polydata = read_vtp_file("/home/sces213/Malaika/lts_modulus/stokes_mgn/results/res_0.vtp")
point_data = polydata.GetPointData()
u = np.array(point_data.GetArray("u"))
v = np.array(point_data.GetArray("v"))
p = np.array(point_data.GetArray("p"))
```

```python
u
```

```python


# Create a KDE plot for each distribution
plt.figure(figsize=(10, 6))

sns.kdeplot(u, label='u', color='blue', linewidth=2)
sns.kdeplot(v, label='v', color='green', linewidth=2)
sns.kdeplot(p, label='p', color='red', linewidth=2)

# Add a Gaussian distribution for comparison (optional)
x_vals = np.linspace(-5, 5, 1000)
plt.plot(x_vals, stats.norm.pdf(x_vals, 0, 1), '--', label='Gaussian (μ=0, σ=1)', color='black')

# Customize the plot
plt.title('Distribution of (u,v,p)')
plt.xlabel('Displacement Value')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()
```
